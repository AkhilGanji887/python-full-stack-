```python
# class attribute - school
# instance attribute - id, name , class
class student:
    school="codegnan"
    def __init__(self,stdid, name, standard):
        self.stdid =stdid # instance attributes
        self.name =name # instance attributes
        self.standard =standard
    def student_details(self):
        return f" Myself {self.name} and I am from {self.standard} class"
 # class method
@classmethod
def update_school(cls, new_school_name):
    cls.school = new_school_name
    return "school name updated"
@staticmethod
def percentage(marks:list):
    return (sum(marks)/300) * 100

    
s1=student(101, "Ravi", 9)
s2=student(102, "Raju", 10)
print(s1.school, s2.school)
print(s1.student_details())
print(s1.update_school("CODEEDE school"))
print(s1.school, s2.school)
print(s2.percentage([85,80,90]))
```

- class method: The method defination starts with 'classmethod' decorator and it takes first parameter as 'cls' those methods called as 'class methods'.This methods works on class Attributes.
- `syntax:`<br>
- `            class class_name:`            <br>
- `                  @classmethod`           <br>
- `                  #blockcode`             <br>
- static method:The method defination starts with `staticmethod` decorator and it doesn't take first parameter as `self or cls those methods called as  `static methods`.This methods works on except instance attributes and class attributes.syntax:

- the intialization constructor is used for to intialiaze the instance attribute data and this constructor takes first parameter as self and this constructor executes whenever an object is created for class


```python
# class attribute - school
# instance attribute - id, name , class

# class attribute - school
# instance attribute - id, name , class
class student:
    school="codegnan"
    def __init__(self,stdid, name, standard):
        self.stdid =stdid # instance attributes
        self.name =name # instance attributes
        self.standard =standard
    def student_details(self):
        return f" Myself {self.name} and I am from {self.standard} class"
 # class method
@classmethod
def update_school(cls, new_school_name):
    cls.school = new_school_name
    return "school name updated"
@staticmethod
def percentage(marks:list):
    return (sum(marks)/300) * 100

    
s1=student(101, "Ravi", 9)
s2=student(102, "Raju", 10)
print(s1.school, s2.school)
print(s1.student_details())
print(s1.update_school("CODEEDE school"))
print(s1.school, s2.school)
print(s2.percentage([85,80,90]))
```
