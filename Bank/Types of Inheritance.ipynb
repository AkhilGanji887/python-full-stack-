# Inheritance

```python
# Multi-Level Inheritance
class Person:
    def __init__(self, name):
        self.name = name
    def display(self):
        return f"Name: {self.name}"
class Doctor(Person):
    def __init__(self, name, specialization):
        super().__init__(name)
        self.specialization = specialization
class Surgeon(Doctor):
    def __init__(self, name, specialization, salary):
        super().__init__(name, specialization)
        self.salary = salary
    def details(self):
        return f"{self.display()}, Specialization: {self.specialization}, Salary: ₹{self.salary}"
s = Surgeon("Rahul", "Cardiology", 90000)
print(s.details())
```

```python
# multiple Inheritance
class Teacher:
    def teach(self):
        print("Teaching Python")
class Coach:
    def train(self):
        print("Training Cricket")
class Mentor(Teacher, Coach):
    def guide(self):
        print("Guiding Students")
m = Mentor()
m.teach()
m.train()
m.guide()
```

```python
# Hierarichial Inheritance
class Vehicle:
    def __init__(self, brand):
        self.brand = brand
class Car(Vehicle):
    def info(self):
        print(f"{self.brand} Car")
class Bike(Vehicle):
    def info(self):
        print(f"{self.brand} Bike")
c = Car("Hyundai")
b = Bike("Yamaha")
c.info()
b.info()
```

```python
# Hybrid Inheritance
class Employee:
    def work(self):
        print("Employee is working")
class Developer(Employee):
    def code(self):
        print("Writing Python code")
class Designer:
    def design(self):
        print("Designing UI")
class TeamMember(Developer, Designer):
    def project(self):
        print("Working on Full Stack Project")
t = TeamMember()
t.work()
t.code()
t.design()
t.project()
```
