### Encapsulation:
- Encapsulation means,wraping the methods and attributes in single unit that one is class.The main aim of Encapsulation is to restrict the direct access of attributes and methods.
     - This type of Restriction can implement by using 'Access Modifiers'
     - Types of Access modifiers
         1. public
         2. protected
         3. private
     -In python We don't have any keywords to represent it public,protected and private,but by using underscore(_) can define.<br>
           1. Public : The public members(i.e attributes + method)doesn't starts with under score.This public member can access anywhere in the program.<br>
           2. Protected: The Public Members(i.e attributes + method)starts with single under score.This members can access in inside the class and in subclasses also.<br>
           3. Private:  The Private Members(i.e attributes + method)starts with double under score.This members can access in only inside.
