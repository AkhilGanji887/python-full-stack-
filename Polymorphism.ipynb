# Polymorphism

> In Python 'Polymorphism' `poly` means `many` and morphism means `forms`.
> In Polymorphism single function its show different behaviours based on input context.
> In Python Polymorphism can achieve through following ways<br>
      1)Polymorphism Through Operators<br>
      2)Polymorphism through Functions<br>
      3)Polymorphism through methods<br>
          a)Runtime Polymorphism (method overriding)<br>
          b)compile time Polymorphism (method overloading)<br>

> 1)Polymorphism Through Operators:<br>
> Here single operator it shows different actions based on input context.<br>
> For Example '+' operator it perform addition operation when inputs are integers as well as it forms string concatenation when inputs are strings.<br>
-                     Eg:  3 + 5 = 8-->Addition
-                          'code'+'gnan'-->string concatenation
-                           3 * 'code'-->multiplication


> 2)Polymorphism Through Functions:<br>
> Here a Function shows different actions based on input context.
-                 Eg:  len('Hello')-->string length
-                      len([10,2,3,4])-->List length

> 3)Polymorphism Through Methods:<br>
> -->(through the inhertance)<br>
> A method defined in child class that is exist in parent class with the same name.whenever if you call that method through child class object,the child class method overrides the parent class method.This is known as method overriding.
-           Eg:class A:
-               def show()
-              class B(A):
-               def show()
-            b = B()
-             b.show()

```python
class person:
    def __init__(self, name,age,gender):
        self.name = name
        self.age = age
        self.gender = gender
    def details(self):
        return f"My name is {self.name} and i am a {self.age} years old"

class student(person):
    def __init__(self, name,age,gender,stdid,branch):
        super().__init__(name, age, gender)
        self.stdid = stdid
        self.branch = branch
    def details(self):
        print(super().details())
        return f"My name is {self.name} and i am belongs to {self.branch} Branch"
s = student("Ravi",20,"M",101, "CSE")
print(s.details())
    
```

```python
# Method Overriding
# Even if you define different number of parameter  in method
#                        always
class Person:
    def __init__(self, stdid, name, age, gender):
        self.stdid = stdid
        self.name = name
        self.age = age
        self.gender = gender
    def details(self):
        return f"My name is {self.name} and I am {self.age} years old."
class Student(Person):
    def __init__(self, stdid, name, age, gender, branch):
        super().__init__(stdid, name, age, gender)
        self.branch = branch
    def details(self):
        return f"My name is {self.name} and I belong to the {self.branch} branch."
s = Student(101, "Ravi", 20, "M", "CSE")
print(s.details())
```

```python
# Method Overloading
> A class contains multiple methods with same name with different parameters list.In that case last defined method overrides the previous 
defined metod.This called as method overloading.
> Directly python doesn't supports method overloading but we can achieve same functionality through default arguements, variables length 
positions,orvariable length keyword arguments
```

```python
# Method overloading
# Directly python dosn't supports method overloading but, we can achieve same function 
#                           through the default arguments or Arbitrary Arguments
class Math:
    def add(self, a):
        return a
    def add(self, a,b):
        return a + b
    def add(self, a,b,c):
        return a + b + c
    def add(self, a,b,c):
        return a + b + c 
    m = Math()
print(m.add(1,2,3))
```

```python
# __str__() magic method
class point:
    def __init__(self, x):
        self.x = x
    def __str__(self):
        return f"This is point class and point value is {self.x}"
    def __add__(self, other):
        return self.x + other.x
p1 = point(5)
print(p1)
p2 = point(3)
print(p1.x+ p2.x)
p3 = p1 + p2
print(p3)
```
