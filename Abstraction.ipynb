# Abstraction

```python
from abc import ABC, abstractmethod
class Mobile(ABC):
    @abstractmethod
    def ram(self):
        pass
    @abstractmethod
    def rom(self):
        pass
    @abstractmethod
    def battery(self):
        pass
    # concrete Method implementation
    def display(self, Hzs):
        return f" This mobile display quality is {Hzs}"
class Samsung(Mobile):
    def __init__(self, model, battery_size, display_quality):
        self.model = model
        self.battery_size = battery_size
        self.display_quality = display_quality
    def ram(self, ram_size):
        return f" The Samsung mobile {self.model} has {ram_size}GB RAM"
    def rom(self, rom_size):
        return f" The Samsung mobile {self.model} has {rom_size}GB ROM"
    def battery(self):
        return f" The Samsung mobile {self.model} has {self.battery_size} mah Battery Capacity"
    def camera(self, primary, secondary):
        return f" The Samsung mobile {self.model} mobile have primary Camera with {primary}MP and secondary camera with {secondary}MP"
#m=Mobile()  TypeError: Can't instantiate abstract class Mobile with abstract methods battery, ram, rom
Samsung = Samsung("Galaxy S26", 5000, 120)
print(Samsung.ram(16))
print(Samsung.rom(1024))
print(Samsung.battery())
print(Samsung.camera(200,50 ))
print(Samsung.display(120))
```

```python
from abc import ABC, abstractmethod
class Mobile(ABC):
    @abstractmethod
    def ram(self):
        pass
    @abstractmethod
    def rom(self):
        pass
    @abstractmethod
    def battery(self):
        pass
    # concrete method
    def display(self, Hzs):
        return f"This mobile display quality is {Hzs}Hz"
class Oppo(Mobile):
    def __init__(self, model, ram_size, rom_size, battery_size, display_quality):
        self.model = model
        self.ram_size = ram_size
        self.rom_size = rom_size
        self.battery_size = battery_size
        self.display_quality = display_quality
    def ram(self):
        return f"The Oppo mobile {self.model} has {self.ram_size}GB RAM"
    def rom(self):
        return f"The Oppo mobile {self.model} has {self.rom_size}GB ROM"  
    def battery(self):
        return f"The Oppo mobile {self.model} has {self.battery_size} mAh Battery Capacity"
    def camera(self, primary, secondary):
        return f"The Oppo mobile {self.model} has a {primary}MP primary camera and a {secondary}MP secondary camera"
# Correct instantiation
reno15 = Oppo("reno15", 4, 128, 5000, 120)
print(reno15.ram())
print(reno15.rom())
print(reno15.battery())
print(reno15.camera(32, 52))
print(reno15.display(reno15.display_quality))
```
